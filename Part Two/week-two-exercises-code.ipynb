```python
from matplotlib import image, pyplot as plt
import numpy as np
from PIL import Image
import urllib

# flag = np.array(Image.open(urllib.request.urlopen("https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRAWQgQstiKa1U6P0QfpcnlRUbIR0IvIDfhAQ&s")))

# plt.imshow(flag)
# plt.show()

# URL of the image
lenna_data = image.imread("lenna.bmp")

url = "https://singifyai.fineshare.com/soundboard-cover/indian-soundboard.webp"
# Create a request object with a User-Agent header
headers = {'User-Agent': 'Chrome/6.0'}
req = urllib.request.Request(url, headers=headers)
# Open the image URL and convert it to a NumPy array
with urllib.request.urlopen(req) as response:
    flag_data = np.array(Image.open(response))

combined_data = lenna_data.copy()
for width in range(256):
    for height in range(256):
        combined_data[height][width + 256] = flag_data[height][width]

image.imsave("combined.bmp", combined_data)

# Display the image using matplotlib
plt.imshow(combined_data)
plt.axis('off')  # Turn off axis labels
plt.show()
```
